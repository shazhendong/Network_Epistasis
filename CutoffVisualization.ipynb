```python
import pandas as pd

# read the network investigation results
file = 'Mouse_XOR2W_filtered.csv_networkInvestigation_pval.csv'
df = pd.read_csv(file)

cutoff_100 = 0.0095
cutoff_99 = 0.0055
cutoff_95 = 0.0014
cutoff_90 = 0.0005

# get the string before the second underscore of file name
title = file.split('_')[0] + '_' + file.split('_')[1]
```

```python
# plot threshold vs. modularity_greedy using matplotlib
# add vertical line at threshold = 0.03
# add vertical line at threshold = 0.035

import matplotlib.pyplot as plt
import numpy as np

plt.plot(df['threshold'], df['modularity_greedy'], '-')
plt.axvline(x=cutoff_100, color='r', linestyle='--') # 100%
plt.axvline(x=cutoff_99, color='b', linestyle='--') # 99% 
plt.axvline(x=cutoff_95, color='g', linestyle='--') # 95%
plt.axvline(x=cutoff_90, color='y', linestyle='--') # 90%
# add legend for the vertical lines
plt.legend(['Modularity', '100%', '99%', '95%', '90%'], loc='lower right')
plt.xlabel('Threshold')
plt.ylabel('Modularity')
plt.title(title)
#plt.show()

# save the plot to pdf named as the title
plt.savefig(title + '.pdf')

```
